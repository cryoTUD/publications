```python
import os 

new_calibration_test_dir = "/home/abharadwaj1/papers/elife_paper/figure_information/archive_data/organized/rebuttal/larger_dataset/testing_calibration"
emdb_ids = [x.split("_")[1] for x in os.listdir(new_calibration_test_dir) if os.path.isdir(os.path.join(new_calibration_test_dir, x))]
emdb_ids
```

```python
import requests
import matplotlib.pyplot as plt
def emdb_resolution(emdb_id):
    url = f"https://www.ebi.ac.uk/emdb/api/entry/processing/{emdb_id}"

    response = requests.get(url)
    if response.status_code == 200:
        resolution = response.json()["processing"]["final_reconstruction"]["final_reconstruction_type"]["resolution"]["valueOf_x"]
        return resolution
    else:
        raise ValueError(f"EMDB ID {emdb_id} not found or API error.")
```

```python
for emdb in emdb_ids:
    try:
        resolution = emdb_resolution(emdb)
        print(f"EMDB ID: {emdb}, Resolution: {resolution} Å")
    except ValueError as e:
        print(e)
```
