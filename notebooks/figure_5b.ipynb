# Figure 5b – Violin plots: helix vs sheet Debye breakpoint frequencies

Reads `data/processed/figure_5b.csv` and saves `figures/figure_5b.png`.


```python
import sys
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns

cwd = Path.cwd()
ROOT_DIR = cwd.parent if cwd.name == 'notebooks' else cwd
sys.path.insert(0, str(ROOT_DIR / 'notebooks'))
from plot_utils import setup_style

PROCESSED = ROOT_DIR / 'data' / 'processed'
FIGURES   = ROOT_DIR / 'figures'
FIGURES.mkdir(parents=True, exist_ok=True)

setup_style(font_scale=1.5)
```

```python
df = pd.read_csv(PROCESSED / 'figure_5b.csv')
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
sns.violinplot(
    data=df,
    x='structure_type', y='breakpoint_A',
    order=['helix', 'sheet'],
    linewidth=3,
    ax=ax,
    palette={'helix': 'blue', 'sheet': 'yellow'},
    alpha=0.5,
)
ax.set_xticklabels(['Helix Profiles', 'Sheet Profiles'])
ax.set_ylabel(r'Resolution [$\AA$]')
ax.set_xlabel('')
ax.set_yticks([3, 4, 5, 6])
plt.tight_layout()
fig.savefig(FIGURES / 'figure_5b.png', dpi=300, bbox_inches='tight', format='png')
print('Saved figures/figure_5b.png')
```
