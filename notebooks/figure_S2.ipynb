# Figure S2 – Wilson B-factor fit quality at specific atom positions

Reads `data/processed/figure_S2.csv` and saves `figures/figure_S2_C67.png` and `figures/figure_S2_A303.png`.


```python
import sys
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib
import warnings
warnings.filterwarnings('ignore')

cwd = Path.cwd()
ROOT_DIR = cwd.parent if cwd.name == 'notebooks' else cwd
sys.path.insert(0, str(ROOT_DIR / 'notebooks'))
from plot_utils import setup_style, pretty_plot_radial_profile

PROCESSED = ROOT_DIR / 'data' / 'processed'
FIGURES   = ROOT_DIR / 'figures'
FIGURES.mkdir(parents=True, exist_ok=True)

setup_style(font_scale=2.0)
```

```python
df    = pd.read_csv(PROCESSED / 'figure_S2.csv')
freq  = df['freq'].values

rp1   = df['rp_C67_CZ'].values
exp1  = df['exp_C67_CZ'].values
q1    = df['qfit_C67_CZ'].iloc[0]

rp2   = df['rp_A303_CA'].values
exp2  = df['exp_A303_CA'].values
q2    = df['qfit_A303_CA'].iloc[0]
```

```python
for tag, rp, exp, qfit in [('C67_CZ', rp1, exp1, q1), ('A303_CA', rp2, exp2, q2)]:
    fig = pretty_plot_radial_profile(
        freq, [exp, rp],
        legends=[f'$R^2$: {qfit}', 'Measured'],
        logScale=True,
        normalise=False,
        squared_amplitudes=True,
        ylims=[-5, 10],
        linewidth=2,
        figsize=(7, 5),
        fontsize=12,
    )
    fname = FIGURES / f'figure_S2_{tag}.png'
    fig.savefig(fname, dpi=300, bbox_inches='tight', format='png')
    print(f'Saved {fname}')
```
