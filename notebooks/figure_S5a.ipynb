# Figure S5a – Theoretical helix / sheet / average profiles

Reads `data/processed/figure_5c_theoretical.csv` and saves `figures/figure_S5a.png`.


```python
import sys
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib
import warnings
warnings.filterwarnings('ignore')

cwd = Path.cwd()
ROOT_DIR = cwd.parent if cwd.name == 'notebooks' else cwd
sys.path.insert(0, str(ROOT_DIR / 'notebooks'))
from plot_utils import setup_style, pretty_plot_radial_profile

PROCESSED = ROOT_DIR / 'data' / 'processed'
FIGURES   = ROOT_DIR / 'figures'
FIGURES.mkdir(parents=True, exist_ok=True)

setup_style(font_scale=1.5)
```

```python
df   = pd.read_csv(PROCESSED / 'figure_5c_theoretical.csv')
freq = df['freq'].values
profiles = [
    df['helix_theoretical'].values,
    df['average_theoretical'].values,
    df['sheet_theoretical'].values,
]
labels = [
    'Average helix profile',
    'Average profile',
    'Average sheet profile',
]
```

```python
fig = pretty_plot_radial_profile(
    freq, profiles,
    legends=labels,
    logScale=False,
    normalise=True,
    squared_amplitudes=True,
    ylims=[-0.001, 0.035],
    yticks=[0, 0.01, 0.02, 0.03],
    linewidth=2,
    figsize=(5, 3),
    fontsize=11,
)
fig.savefig(FIGURES / 'figure_S5a.png', dpi=300, bbox_inches='tight', format='png')
print('Saved figures/figure_S5a.png')
```
