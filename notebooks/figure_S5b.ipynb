# Figure S5b – FSC between repaired maps (helix vs sheet, helix vs average)

Reads `data/processed/figure_S5b_fsc.csv` and saves `figures/figure_S5b.png`.


```python
import sys
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib
import warnings
warnings.filterwarnings('ignore')
cwd = Path.cwd()
ROOT_DIR = cwd.parent if cwd.name == 'notebooks' else cwd
sys.path.insert(0, str(ROOT_DIR / 'notebooks'))
from plot_utils import setup_style, plot_fsc_curves
from new_plot_utils import pretty_lineplot_multiple_fsc_curves

PROCESSED = ROOT_DIR / 'data' / 'processed'
FIGURES   = ROOT_DIR / 'figures'
FIGURES.mkdir(parents=True, exist_ok=True)

setup_style(font_scale=2.0)
```

```python
df   = pd.read_csv(PROCESSED / 'figure_S5b_fsc.csv')
freq = df['freq'].values
fsc_d = {
    'helix_vs_sheet':   df['fsc_helix_vs_sheet'].values,
    'helix_vs_average': df['fsc_helix_vs_average'].values,
}
labels = {
    'helix_vs_sheet':   r'Using all-$\alpha$-helix v all-$\beta$-sheet',
    'helix_vs_average': r'Using all-$\alpha$-helix v mixed $\alpha$$\beta$',
}
```

```python
fig = plot_fsc_curves(freq, fsc_d, labels=labels, linewidth=2, figsize=(7, 5), fontsize=12)
fig.savefig(FIGURES / 'figure_S5b.png', dpi=300, bbox_inches='tight', format='png')
print('Saved figures/figure_S5b.png')
```
