```python
from pathlib import Path

import os 
import subprocess


# ── resolve project root ────────────────────────────────────────────────────

os.environ["ElectronScattering2022_ROOT"] = os.getcwd()

ROOT_DIR = Path(os.environ["ElectronScattering2022_ROOT"])
SCRIPT_PATH = ROOT_DIR / "print_all_figures.py"

```

```python
# Download data
import os 
import wget 
import zipfile
data_folder = os.path.join(ROOT_DIR, "data")
if os.path.exists(data_folder):
    print("Data already exists. Skipping download.")

url_for_plot_data = "https://surfdrive.surf.nl/public.php/dav/files/c6LzjA5T8MW7bSq/?accept=zip"
# download the data
PLOT_DATA_STORE_PATH_zip = data_folder + ".zip"
wget.download(url_for_plot_data, out=PLOT_DATA_STORE_PATH_zip)

with zipfile.ZipFile(PLOT_DATA_STORE_PATH_zip, 'r') as zip_ref:
    zip_ref.extractall(ROOT_DIR)

os.remove(PLOT_DATA_STORE_PATH_zip)
if os.path.exists(data_folder):
    print("Data downloaded and extracted successfully.")
else:
    raise Exception("Data download or extraction failed.")

```

```python
cmd = ["python3"]
cmd.append(str(SCRIPT_PATH))


subprocess.run(cmd, stderr=subprocess.DEVNULL)
```
