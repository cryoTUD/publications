# Figure 2f – Local atomic B-factor vs Wilson B-factor scatter plot

Reads `data/processed/figure_2f.csv` and saves `figures/figure_2f.png`.


```python
import sys
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns

cwd = Path.cwd()
ROOT_DIR = cwd.parent if cwd.name == 'notebooks' else cwd
sys.path.insert(0, str(ROOT_DIR / 'notebooks'))
from plot_utils import setup_style, plot_correlations

PROCESSED = ROOT_DIR / 'data' / 'processed'
FIGURES   = ROOT_DIR / 'figures'
FIGURES.mkdir(parents=True, exist_ok=True)

setup_style(font_scale=2.0)
```

```python
df = pd.read_csv(PROCESSED / 'figure_2f.csv')
# Apply same filter as original: keep atomic_bfactor > 40
df_filt = df[df['atomic_bfactor'] > 40].copy()
# print(f'Filtered samples: {len(df_filt)} / {len(df)}')
```

```python
fig = plot_correlations(
    df_filt['atomic_bfactor'], df_filt['wilson_bfactor'], 
    scatter=True, 
    figsize=(10, 10),
    x_label="Local Atomic B-factors $\AA^{2}$",
    y_label="Local Wilson B-factors $\AA^{2}$",
)
fig.savefig(FIGURES / 'figure_2f.png', dpi=300, bbox_inches='tight', format='png')
print('Saved figures/figure_2f.png')

    
```
