# Figure 1b – Radial profiles of backbone-only maps (uniform B-factors)

Reads `data/processed/figure_1b.csv` and saves `figures/figure_1b.png`.


```python
import sys
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib
import warnings
warnings.filterwarnings('ignore')
cwd = Path.cwd()
ROOT_DIR = cwd.parent if cwd.name == 'notebooks' else cwd
sys.path.insert(0, str(ROOT_DIR / 'notebooks'))

from plot_utils import setup_style, pretty_plot_radial_profile

PROCESSED = ROOT_DIR / 'data' / 'processed'
FIGURES   = ROOT_DIR / 'figures'
FIGURES.mkdir(parents=True, exist_ok=True)

setup_style(font_scale=2.0)
```

```python
df = pd.read_csv(PROCESSED / 'figure_1b.csv')
freq         = df['freq'].values
bfactor_cols = [c for c in df.columns if c.startswith('bfactor_')]
profiles     = [df[c].values for c in bfactor_cols]
```

```python
fig = pretty_plot_radial_profile(
    freq, 
    profiles,
    showlegend=False, 
    normalise=True, 
    squared_amplitudes=False, 
    logScale=False, 
    ylims=[-0.005,0.2],
    crop_freq=[100,2], 
    linewidth=2, 
    fontscale=4
)
fig.savefig(FIGURES / 'figure_1b.png', dpi=300, bbox_inches='tight', format='png')
print('Saved figures/figure_1b.png')
```
